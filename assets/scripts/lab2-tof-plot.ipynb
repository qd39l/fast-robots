```python
import matplotlib.pyplot as plt

ref_distance =       [100, 300, 500, 700, 1000, 1500, 1700, 2000]
short_distance_box = [103, 307, 700, 973, 1400, 300,  400, 600]
long_distance_box =  [104, 307, 700, 970, 1398, 1400, 900, 600]

fig, ax = plt.subplots()
fig.set_dpi(150)
ax.plot(ref_distance, short_distance_box, label='Short Mode')
ax.plot(ref_distance, long_distance_box, label='Long Mode')
ax.set_title('SparkFun VL53L1X Profile with a Box')
ax.set_xlabel('reference distance (mm)')
ax.set_ylabel('measured distance (mm)')
ax.legend()

# Long mode ranging time = 100 ms
# Short mode ranging time = 50 ms
```
