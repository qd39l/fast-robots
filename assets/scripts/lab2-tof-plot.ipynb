```python
import matplotlib.pyplot as plt

ref_distance =       [100, 300, 500, 700, 900, 1000, 1300, 1500, 1700, 2000]
short_distance_box = [98,307,512,725,937,1040,1360,1560,1750,2040]
long_distance_box =  [100,310,512,722,928,1040,1330,1550,1750,2050]

fig, ax = plt.subplots()
fig.set_dpi(150)
ax.plot(ref_distance, ref_distance, label='ref distance')
ax.plot(ref_distance, short_distance_box, label='Short Mode')
ax.plot(ref_distance, long_distance_box, label='Long Mode')
ax.set_title('SparkFun VL53L1X Distance Modes')
ax.set_xlabel('reference distance (mm)')
ax.set_ylabel('measured distance (mm)')
ax.legend()
```
